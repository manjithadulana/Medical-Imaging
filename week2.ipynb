```python
import itk
import matplotlib.pyplot as plt
import numpy as np
from pandas import array

input_fn =  'images/image-000001.dcm'
output_fn = 'images/image_written.dcm'

image = itk.imread(input_fn)

assert image is not None  # import this 

print(image.shape)

t1 = [0 for i in range(100)]
t2 = [ 0.7*i + 80 for i in range(100,200)]
t3 = [255 for i in range(56)]

t = np.concatenate((t1,t2,t3) , axis= 0)
result = t[image]



fig,ax =plt.subplots(1,2, figsize=(10,20))
ax[0].imshow(image[0], cmap='gray')
ax[1].imshow(result[0], cmap='gray')
plt.show()
```
