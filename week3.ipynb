## Filtering

```python
import itk
import matplotlib.pyplot as plt

input_filename ='images/image-000001.dcm'
image_orignal =  itk.imread(input_filename)
image =  image_orignal.astype(itk.F)

smoothed =  itk.anti_alias_binary_image_filter(image,number_of_iterations=3)

# smoother = itk.AntiAliasBinaryImageFilter.New(image,NumberofIterations=3)
# smoother.update()
# smoothed =smoother.GetOutput()

smoother = itk.AntiAliasBinaryImageFilter.New(image)
smoother.SetNumberOfIterations(3)
smoother.Update()
smoothed =smoother.GetOutput()

fig,ax = plt.subplots(1,2,figsize=(10,20))
ax[0].imshow(image_orignal[0])
ax[1].imshow(smoothed[0])
plt.show()
```

## Binary Thresholding

```python
import itk
from matplotlib.figure import figaspect
import matplotlib.pyplot as plt

input_image = 'images/BrainProtonDensitySlice.png'
output_image = 'images/binarythreshold.png'

lower_threshold,upper_threshold, outside_value,inside_value = 160,200,0,255

PixelType = itk.UC
Dimension =2
ImageType = itk.Image[PixelType,PixelType]

reader = itk.ImageFileReader[ImageType].New()
reader.SetFileName(input_image)

thresholdFilter = itk.BinaryThresholdImageFilter(ImageType,ImageType).New()
thresholdFilter.SetInput(reader.GetOutput())

thresholdFilter.SetLowerThreshold(lower_threshold)
thresholdFilter.SetUpperThreshold(upper_threshold)
thresholdFilter.SetOutsideValue(outside_value)
thresholdFilter.SetInsideValue(inside_value)

writer = itk.ImageFileWriter[ImageType].New()
writer.SetFileName(output_image)
writer.SetInput(thresholdFilter.GetOutput())
writer.Update()

fig,ax = plt.subplot(1,2, figsize=(10,20))
ax[0].imshow(reader.GetOutput())
ax[1].imshow(thresholdFilter.GetOutput())
```
